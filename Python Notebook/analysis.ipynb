```python
sum_cue=[0,0,0,0,0,0] # 0.0, 0.05, 0.1, 0.2, 0.3, 0.5
sum_un=[0,0,0,0,0,0]
sum_cen=[0,0,0,0,0,0]
ctr=0
ctr_cue=[0,0,0,0,0,0] # 0.0, 0.05, 0.1, 0.2, 0.3, 0.5
ctr_un=[0,0,0,0,0,0]
ctr_cen=[0,0,0,0,0,0]
ctr_n=0

for index, row in df.iterrows():
    if ((index >1) & (index <122)):
        cue=0
        targ=0
        j=-1
        strt=row['startTime']
        if(strt==0.0):
            j=0
        if(strt==0.05):
            j=1
        if(strt==0.1):
            j=2
        if(strt==0.2):
            j=3
        if(strt==0.3):
            j=4
        if(strt==0.5):
            j=5

        if (row['image_cue_1']=='images/light_grey_box_with_cue'):
            if(row['image_target_1']=='images/light_grey_box_with_target'):
                sum_cue[j]+=row['keyResponse.rt']
                ctr_cue[j]+=1

            elif (row['image_target_3']=='images/light_grey_box_with_target'):
                sum_un[j]+=row['keyResponse.rt']
                ctr_un[j]+=1

            elif (row['image_target_2']=='images/light_grey_box_with_target'):
                sum_cen[j]+=row['keyResponse.rt']
                ctr_cen[j]+=1
            else:
                ctr_n+=1

        elif (row['image_cue_3']=='images/light_grey_box_with_cue'):
                if(row['image_target_3']=='images/light_grey_box_with_target'):
                    sum_cue[j]+=row['keyResponse.rt']
                    ctr_cue[j]+=1

                elif (row['image_target_1']=='images/light_grey_box_with_target'):
                    sum_un[j]+=row['keyResponse.rt']
                    ctr_un[j]+=1

                elif (row['image_target_2']=='images/light_grey_box_with_target'):
                    sum_cen[j]+=row['keyResponse.rt']
                    ctr_cen[j]+=1
                else:
                    ctr_n+=1
        else:
            if(row['image_target_1']=='images/light_grey_box_with_target'):
                sum_un[j]+=row['keyResponse.rt']
                ctr_un[j]+=1
                
            elif (row['image_target_3']=='images/light_grey_box_with_target'):
                sum_un[j]+=row['keyResponse.rt']
                ctr_un[j]+=1
                
            elif (row['image_target_2']=='images/light_grey_box_with_target'):
                sum_cen[j]+=row['keyResponse.rt']
                ctr_cen[j]+=1
            else:
                ctr_n+=1

        

print(sum_cue)
print(sum_un)
print(sum_cen)
print(ctr_cue)
print(ctr_cen)
print(ctr_un)
print(ctr_n)
```
